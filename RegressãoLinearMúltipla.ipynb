```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
df=pd.read_csv("/content/insurance.csv", delimiter=';')
```

```python
df
```

```python
x=df.iloc[:,:-1].values
y= df.iloc[:,-1].values
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder_X = LabelEncoder() #transformando letras em nÃºmeros binÃ¡rios
```

```python
x[:, 3] = labelencoder_X.fit_transform(x[:, 3]) #atribundo o valor binÃ¡rio a coluna 3 (regiÃ£o)
```

```python
d=pd.get_dummies(x[:,3]).values
```

```python
x = x[:,:3]
```

```python
for j in range(0, d.shape[1]):
  x = np.insert(x, 3, d[:,j], axis=1)
```

```python
x = x[:,:-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
xTrain, xTest, yTrain, yTest= train_test_split(x,y, test_size=0.2)
```

```python
print (len(xTrain), len(xTest))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor= LinearRegression()
```

```python
regressor.fit(xTrain, yTrain)
```

```python
yPred=regressor.predict(xTest)
```

```python
#mostrando o diferenÃ§a entre o que Ã© do dadaset e o que o modelo aprendeu
#verificando a quantidade de erro
for i in range(0, yTest.shape[0]) :
  print (yPred[i], yTest[i], abs(yPred[i]-yTest[i])) #yPred Ã© o que o modelo disse e yTest Ã© o que estÃ¡ na base
```

```python
prev= regressor.predict(xTest)

from sklearn import metrics
print ("RMSE", np.sqrt (metrics.mean_squared_error(yTest, prev))) #avaliando o erro mÃ©dio da regressÃ£o
```
